```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['input1'] = [1,1,0,0]
or_data['input2'] = [1,0,1,0]
or_data['output'] = [1,1,1,0]
```

```python
and_data['input1'] = [1,1,0,0]
and_data['input2'] = [1,0,1,0]
and_data['output'] = [1,0,0,0]
```

```python
xor_data['input1'] = [1,1,0,0]
xor_data['input2'] = [1,0,1,0]
xor_data['output'] = [0,1,1,0]
```

```python
or_data
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='input1',y='input2',hue='output',data=or_data)
```

```python
and_data
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='input1',y='input2',hue='output',data=and_data)
```

```python
xor_data
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='input1',y='input2',hue='output',data=xor_data)
```

```python
from sklearn.linear_model import Perceptron

or_perceptron = Perceptron()
and_perceptron = Perceptron()
xor_perceptron = Perceptron()
```

```python
or_perceptron.fit(or_data[['input1','input2']],or_data['output'])
and_perceptron.fit(and_data[['input1','input2']],and_data['output'])
xor_perceptron.fit(xor_data[['input1','input2']],xor_data['output'])
```

```python
or_perceptron.coef_ , or_perceptron.intercept_
```

```python
x = np.linspace(-1,1,5)
y = -x+0.5
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
sns.scatterplot(x='input1',y='input2',hue='output',data=or_data)
```

```python
x = np.linspace(-1,1,5)
y = -x+1
plt.figure(figsize=(5,3))
plt.plot(x,y)
sns.scatterplot(x='input1',y='input2',hue='output',data=and_data)
```

```python
xor_perceptron.coef_ , xor_perceptron.intercept_
```

```python
plt.figure(figsize=(5,3))
plot_decision_regions(xor_data[['input1','input2']].values,xor_data['output'].values,clf=xor_perceptron)
```

**For whatever number of time you run the perceptron on xor data you wont be able to find the decision boundary.**
