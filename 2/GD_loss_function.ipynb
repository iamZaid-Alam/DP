```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, random_state=41, class_sep=15, hypercube=False,
                           n_clusters_per_class=1, n_classes=2)
print(X.shape, y.shape)
```

```python
plt.figure(figsize=(10,4))
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter')
plt.show()
```

```python
def perceptron(X,y):

  w1=w2=b=1
  lr = 0.1

  for i in range(1000):

    for i in range(X.shape[0]):

      z = w1*X[i][0] + w2*X[i][1] + b

      if z*y[i] < 0:
        w1 = w1 + lr*y[i]*X[i][0]
        w2 = w2 + lr*y[i]*X[i][1]
        b  = b + lr*y[i]

  return w1,w2,b
```

```python
w1,w2,b = perceptron(X,y)
```

```python
print(w1,w2,b)
```

```python
m = -(w1/w2)

c = -(b/w2)

print(m,c)
```

```python
x_input = np.linspace(-10,10,100)
y_input = m*x_input + c

plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,'r',linewidth=3)
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter',s=100)
plt.ylim(-3,2)
```
